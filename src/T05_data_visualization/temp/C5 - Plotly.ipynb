```python
# ...existing code...
# Basic scatter plot
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [1, 4, 9, 16, 25],
    "group": ["A", "A", "B", "B", "B"]
})

fig = px.scatter(df, x="x", y="y", color="group", size="y", title="Scatter with size & color")
fig.show()
# ...existing code...
# Line chart with hover and markers
import numpy as np

t = np.linspace(0, 2*np.pi, 200)
df_line = pd.DataFrame({"t": t, "sin": np.sin(t), "cos": np.cos(t)})

fig = px.line(df_line, x="t", y=["sin", "cos"], markers=True, title="Sine & Cosine")
fig.update_traces(mode="lines+markers", hovertemplate="%{y:.3f}")
fig.show()
# ...existing code...
# Bar chart with facetting
df_bar = pd.DataFrame({
    "category": ["A", "A", "B", "B"] * 2,
    "sub": ["X", "Y"] * 4,
    "value": [4, 7, 1, 5, 3, 6, 2, 4]
})

fig = px.bar(df_bar, x="sub", y="value", color="sub", facet_col="category",
             title="Faceted bar chart")
fig.show()
# ...existing code...
# Histogram with marginal box
df_hist = pd.DataFrame({"value": np.random.normal(loc=0, scale=1, size=500)})

fig = px.histogram(df_hist, x="value", nbins=30, marginal="box", title="Histogram with marginal box")
fig.show()
# ...existing code...
# Subplots with make_subplots + go
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=2, subplot_titles=("Scatter", "Bar"))
fig.add_trace(go.Scatter(x=df["x"], y=df["y"], mode="markers", name="scatter"), row=1, col=1)
fig.add_trace(go.Bar(x=df_bar["sub"], y=df_bar["value"], name="bar"), row=1, col=2)
fig.update_layout(title="Mixed subplots", showlegend=False)
fig.show()
# ...existing code...
```
