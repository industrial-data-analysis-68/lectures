```python
# Removing customer-related columns
sales = sales.drop(columns=["Customer Age", "Customer Gender"])

# Removing financial columns
sales = sales.drop(columns=["Year", "Cost", "Revenue"])
```

```python
# Inject some missing data rows for studetn practice
import numpy as np

rows_missing = np.random.choice(sales.index, size=10, replace=False)
sales.loc[rows_missing, "Date"] = np.nan
```

```python
customers = customers.drop(columns=["ID"])
```

```python
customers
```

```python
# Random customer ages with reasonale distribution
np.random.seed(42)
customers["Age"] = np.random.normal(loc=35, scale=10, size=len(customers)).astype(int)
customers["Age"] = customers["Age"].clip(lower=18, upper=70)
```

```python
# Random customer genders
genders = ["Male", "Female"]
customers["Gender"] = np.random.choice(
    genders, size=len(customers), replace=True, p=[0.7, 0.3]
)
```

```python
# Fill in missing genders
np.random.seed(24)
missing_gender_indices = np.random.choice(customers.index, size=125, replace=False)
customers.loc[missing_gender_indices, "Gender"] = np.nan
```

```python
# Removing some rows to simulate missing data when merging
customers = customers.drop(index=np.random.choice(customers.index, size=50, replace=False))
```

```python
sales
```

```python
dfm = sales.merge(customers, left_on="Customer ID", right_on="Customer ID", how="left")
```
