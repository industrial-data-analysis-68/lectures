## Handling Duplicated Data


```python
import pandas as pd
```

## Primer


```python
# Create a sample DataFrame with duplicated rows
df = pd.DataFrame(
    {
        "Name": ["John", "Rock", "Neil", "Tom", "John", "John"],
        "Rank": [1, 2, 3, 6, 1, 1],
        "Marks": [98, 59, 97, 99, 98, 98],
    }
)
display(df)
```

```python
# Identify duplicated rows, keeping the first occurrence
df.duplicated(keep="first")
```

```python
# Identify duplicated rows, keeping the last occurrence
df.duplicated(keep="last")
```

```python
# Count total number of duplicated rows
df.duplicated().sum()
```

```python
# Display duplicated rows
# Note: we use boolean indexing to filter the DataFrame. We will learn more about this in the next lesson.
df.loc[df.duplicated()]
```

```python
# Remove duplicated rows, keeping the first occurrence
df.drop_duplicates(keep="first")
```

```python
# Remove duplicated rows, keeping the last occurrence
df.drop_duplicates(keep="last")
```

## Real Data


```python
# Load the Adult dataset
adult = pd.read_csv("./data/adult.csv")
```

```python
# Display the first few rows of the dataset
adult.head()
```

```python
# Display dataset information
adult.info()
```

```python
# Identify duplicated rows in the Adult dataset
adult.duplicated()
```

```python
# Count total number of duplicated rows in the Adult dataset
adult.duplicated().sum()
```

```python
# Display duplicated rows in the Adult dataset
adult.loc[adult.duplicated()]
```

```python
# Remove duplicated rows, keeping the first occurrence
adult_dd = adult.drop_duplicates(keep="first")
```

```python
# Display the shape of the original and cleaned datasets
print(adult.shape)
print(adult_dd.shape)
```

```python
# Remove duplicates based on specific columns: "age" and "workclass"
adult_dd2 = adult.drop_duplicates(subset=["age", "workclass"])
print(adult_dd2.shape)
```
