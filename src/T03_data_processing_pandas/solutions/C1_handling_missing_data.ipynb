# Dealing with missing data


```python
import pandas as pd
import numpy as np
```

```python
# Create a DataFrame with some NaN values
data = {
    "A": [1.0, 5.0, 10.0],
    "B": [2.0, 6.0, 11.0],
    "C": [3.0, np.nan, 12.0],
    "D": [4.0, 8.0, np.nan],
}
```

```python
# Create the DataFrame
df = pd.DataFrame(data)
display(df)
```

```python
# Check for missing values in each column
df.isnull().sum()
```

## Eliminating samples or features with missing values


```python
# Drop rows with any missing values
df.dropna()
```

```python
# Drop columns with any missing values
df.dropna(axis=1)
```

```python
# Introduce a NaN value for demonstration
df.loc[3] = np.nan
display(df)
```

```python
# Drop rows where all elements are NaN
df = df.dropna(how="all")
display(df)
```

```python
# Drop rows with less than 4 non-NaN values
df.dropna(thresh=4)
```

```python
# Drop rows with missing values in column 'C'
df.dropna(subset=["C"])
```

# Filling in missing data


```python
# Display the final DataFrame
display(df)
```

```python
# Fill missing values with 0
dft = df.fillna(0)
display(dft)
```

```python
# Fill missing values with -1
dft = df.fillna(-1)
display(dft)
```

```python
# Show mean of each column, ignoring NaN values
df.mean()
```

```python
# Fill missing values with the mean of each column. This is called mean imputation.
dft = df.fillna(df.mean())
display(dft)
```
